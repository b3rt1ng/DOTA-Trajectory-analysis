```python
arr = np.loadtxt("./coordinates/coord_test.csv", comments='#', delimiter=',',skiprows=1, dtype=float)

a = pd.read_csv('./coordinates/coord_test.csv')
#print(a)


traj_p1 = a[['x0','y0']].tail(-1)
#traj_p1 = a[['x0','y0']].head(200).tail(-1)

#print("test ", traj_p1['x0'])

#df['d'] = df.apply(lambda x: some_func(a = x['a'], b = x['b'], c = x['c']), axis=1)

point_maker = lambda x: Point(x['x0'], x['y0'])

traj_p1['points'] = traj_p1.apply(point_maker, axis=1)
```
